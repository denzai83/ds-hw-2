Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
df = table[0]
df
```

```python
x_data=(df.columns[1:])
y_data=df.iloc[27][1:]
plt.plot(
    x_data, y_data, marker='.',
    linestyle='-' )
plt.xlabel('Роки')
plt.ylabel('Коефіцієнт народжуваності в Україні')
plt.title('Коефіцієнт народжуваності в Україні (1950-2019)')
plt.grid()
plt.show()
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.nan, inplace=True)
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові

```python
df['2014'] = df['2014'].astype(float)
df['2019'] = df['2019'].astype(float)
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print('Частка пропусків в колонці:')
for column in df.columns:
    print(f'{column}  {((df[column].isnull().sum()/ len(df[column])) * 100).round(2)}%')
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(index=df.index[-1])
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df.index=[
'Крим',
'Вінницька',
'Волинська',
'Дніпропетровська',
'Донецька',
'Житомирська',
'Закарпатська',
'Запорізька',
'Івано-Франківська',
'Київська',
'Кіровоградська',
'Луганська',
'Львівська',
'Миколаївська',
'Одеська',
'Полтавська',
'Рівненська',
'Сумська',
'Тернопільська',
'Харківська',
'Херсонська',
'Хмельницька',
'Черкаська',
'Чернівецька',
'Чернігівська',
'Київ',
'Севастополь']
df = df.drop('Регіон', axis=1)
df = df.fillna((df.mean()).round(1))
df
```

```python
sns.heatmap(df, annot=True)
plt.title('Теплова карта коефіцієнту народжуваності в регіонах України (1950-2019)')
plt.xlabel('Рік')
plt.ylabel('Регіон')
plt.show()
```

```python
plt.contourf(df.columns, df.index, df, levels=10, cmap='viridis')
plt.colorbar(label='Значення')
plt.xlabel('Рік')
plt.ylabel('Регіон')
plt.title('Графік контурної карти коефіцієнту народжуваності в регіонах України (1950-2019)')
plt.show()
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df['2019'][df['2019'] > df['2019'].mean()]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df['2014'][df['2014'] == df['2014'].max()]
```
